## Splitting a Data into features and target

```python
from sklearn.model_selection import train_test_split, cross_val_score
```

```python
X = df1.iloc[:,0:-1]
Y = df1.iloc[:,-1]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,Y, test_size = 0.3)
```

```python
X_train.shape, y_train.shape, X_test.shape, y_test.shape
```

```python
from sklearn import svm
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score, precision_score, recall_score, confusion_matrix
```

```python
import warnings
warnings.filterwarnings('ignore')
```

# Kernel Tricks

### Linear Kernel

```python
clf = SVC()
param_grid = [{'kernel':['rbf'],'gamma':[60,50,10,5,2,0.9,0.5],'C':[20,15,14,13,12,11,10,0.1,0.001] }] # Gamma is for RBF and C is for polynomial kernal
gsv = GridSearchCV(clf,param_grid,cv=10) # Grid Search CV gives best parameter
gsv.fit(X_train,y_train)
```

```python
gsv.best_params_ , gsv.best_score_ 
```

```python
clf = SVC(C= 20, gamma = 60)
clf.fit(X_train , y_train)
y_pred = clf.predict(X_test)
acc = accuracy_score(y_test, y_pred) * 100
print("Accuracy =", acc)
confusion_matrix(y_test, y_pred)
```

### Polynomial kernel

```python
clf = SVC()
param_grid = [{'kernel':['poly'],'gamma':[60,50,5,10,5,2,0.9,0.5],'C':[15,14,13,12,11,10,0.1,0.001] }] # Gamma is for RBF and C is for polynomial kernal
gsv = GridSearchCV(clf,param_grid,cv=10) # Grid Search CV gives best parameter
gsv.fit(X_train,y_train)
```

```python
gsv.best_params_ , gsv.best_score_ 
```

```python
clf = SVC(C= 15, gamma = 60)
clf.fit(X_train , y_train)
y_pred = clf.predict(X_test)
acc = accuracy_score(y_test, y_pred) * 100
print("Accuracy =", acc)
confusion_matrix(y_test, y_pred)
```

### RBF Kernel

```python
clf = SVC()
param_grid = [{'kernel':['rbf'],'gamma':[50,5,10,5,2,0.9,0.5],'C':[15,14,13,12,11,10,0.1,0.001] }] # Gamma is for RBF and C is for polynomial kernal
gsv = GridSearchCV(clf,param_grid,cv=10) # Grid Search CV gives best parameter
gsv.fit(X_train,y_train)
```

```python
gsv.best_params_ , gsv.best_score_ 
```

```python
clf = SVC(C= 15, gamma = 0.5)
clf.fit(X_train , y_train)
y_pred = clf.predict(X_test)
acc = accuracy_score(y_test, y_pred) * 100
print("Accuracy =", acc)
confusion_matrix(y_test, y_pred)
```
